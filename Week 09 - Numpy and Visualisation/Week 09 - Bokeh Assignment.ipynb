```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()

# Q1. How can you create a Bokeh plot using Python code?

from bokeh.plotting import figure, show

plot = figure(title='My Bokeh Plot', x_axis_label='X-axis', y_axis_label='Y-axis')
plot.circle([1, 2, 3, 4, 5], [2, 4, 1, 3, 5], size=10, color='blue')

show(plot)
```

```python
# Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

# Glyphs are visual markers that represent data points on a plot.

from bokeh.plotting import figure, show

plot = figure(title='My Bokeh Plot', x_axis_label='X-axis', y_axis_label='Y-axis')
plot.circle([9, 2, 7, 4, 1], [5, 1, 7, 9, 6], size=10, color='blue')

show(plot)
```

```python
# Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

from bokeh.plotting import figure, show

plot = figure(title='My Bokeh Plot', x_axis_label='X-axis', y_axis_label='Y-axis')

plot.xaxis.axis_label_text_color = 'brown'
plot.yaxis.axis_label_text_color = 'brown'
plot.xaxis.minor_tick_line_color = 'green'
plot.yaxis.minor_tick_line_color = 'green'
plot.xaxis.axis_label_text_font_size = '12pt'
plot.yaxis.axis_label_text_font_size = '12pt'

plot.title.text_font_size = '14pt'
plot.title.text_color = 'green'

plot.circle([2, 4, 6, 8], [6, 8, 10, 12], legend_label="Data Points")

plot.legend.title = 'Legend'
plot.legend.label_text_font_size = '10pt'
plot.legend.location = 'top_left'

show(plot)
```

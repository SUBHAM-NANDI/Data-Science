```python
# Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

# The 5 plots are as follows

# [1] Scatter Plot : It is used visualize the relationship between two continuous variables. 
# [2] Bar Plot : It is used to display and compare the distribution of categorical data.
# [3] Histogram : It is used to illustrate the distribution of a single numeric variable. 
# [4] Box Plot : It is used to summarize the distribution of a continuous variable through quartiles.
# [5] Violin Plot : It is used to combine a box plot and a KDE plot to represent the distribution of a numeric variable across different categories.
```

```python
# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using 
# x = "timepoint" and y = "signal" for different events and regions.
# Note: timepoint, signal, event, and region are columns in the fmri dataset.

import seaborn as sns
import matplotlib.pyplot as plt

fmri_dataset = sns.load_dataset('fmri')

sns.lineplot( x = "timepoint", y = "signal", data = fmri_dataset, style = "region", hue = "event"  )

plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.title("fMRI Signal Over Time")

plt.show()
```

```python
# ***Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using 
# x = 'pclass', y = 'age' and y = 'fare'.
# Note: pclass, age, and fare are columns in the titanic dataset.

import seaborn as sns
import matplotlib.pyplot as plt

titanic_dataset = sns.load_dataset('titanic')

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic_dataset)
plt.xlabel('Pclass')
plt.ylabel('Age')
plt.title('Age Distribution by Passenger Class')

plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic_dataset)
plt.xlabel('Pclass')
plt.ylabel('Fare')
plt.title('Fare Distribution by Passenger Class')

plt.tight_layout()

plt.show()
```

```python
# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
# parameter for the 'cut' column of the diamonds dataset.

import seaborn as sns
import matplotlib.pyplot as plt

diamond_dataset = sns.load_dataset("diamonds")

sns.histplot(x = "price", data = diamond_dataset, hue = "cut")

plt.xlabel("Price")
plt.ylabel("Count")
plt.title("Diamond prices")

plt.show()
```

```python
# Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
# of the iris dataset.

import seaborn as sns
import matplotlib.pyplot as plt

iris_dataset = sns.load_dataset("iris")

sns.pairplot(data = iris_dataset, hue = "species")

plt.show()
```

```python
# ***Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

import seaborn as sns
import matplotlib.pyplot as plt


flights_dataset = sns.load_dataset('flights')

flights_matrix = flights_dataset.pivot("month", "year", "passengers") # Pivot the dataset to create a matrix

sns.heatmap(flights_matrix, cmap='YlGnBu')

plt.xlabel('Year')
plt.ylabel('Month')
plt.title('Passenger Count Heatmap')

plt.show()
```

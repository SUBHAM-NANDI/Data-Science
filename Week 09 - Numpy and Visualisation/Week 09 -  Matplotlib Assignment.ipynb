```python
#2 Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

#Note: Also add title, xlabel, and ylabel to the plot.

# Scatter plot is used to to visualize data trend by which can understand the trend of data efficiently. It is used to display
# the relationship between two variables in which the x and y coordinates contains two different types of variables.

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Sactter Plot")
plt.scatter(x,y)
plt.show()
```

```python
# *** Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
# Use the following data:
"""import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])"""

# In matplotlob , subplot() fuction is used to create multiple subplot/visulisations/graphs within a single figure.

import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create subplot
fig, axes = plt.subplots(2, 2, figsize=(10, 6))

axes[0, 0].plot(x1, y1)
axes[0, 0].set_title("Line 1")
axes[0, 0].set_xlabel("X-axis")
axes[0, 0].set_ylabel("Y-axis")

axes[0, 1].plot(x2, y2)
axes[0, 1].set_title("Line 2")
axes[0, 1].set_xlabel("X-axis")
axes[0, 1].set_ylabel("Y-axis")

axes[1, 0].plot(x3, y3)
axes[1, 0].set_title("Line 3")
axes[1, 0].set_xlabel("X-axis")
axes[1, 0].set_ylabel("Y-axis")

axes[1, 1].plot(x4, y4)
axes[1, 1].set_title("Line 4")
axes[1, 1].set_xlabel("X-axis")
axes[1, 1].set_ylabel("Y-axis")

# Adjust spacing between subplot
plt.tight_layout()


plt.show()
```

```python
# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Barplot is also one type of visualisation in matplotlib which gives data in categorical format. 
# Barplot cosists of bar graphs where the length of the bars denotes frequency.

# Bar plot
plt.bar(company,profit)
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Sales")
plt.show()

# Hirizontal Bar plot
plt.barh(company,profit)
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Sales")
plt.show()
```

```python
# *** Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

# Boxplot is used get visuaisation of statistical dataset which includes first, second, third quartile and 
# aslo maximum values.

import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Box plot
data = [box1, box2]

plt.boxplot(data)
plt.title("Box Plot")
plt.xlabel("Box")
plt.ylabel("Value")
plt.show()
```

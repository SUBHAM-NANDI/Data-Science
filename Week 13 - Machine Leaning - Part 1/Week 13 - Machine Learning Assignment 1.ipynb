```python
# Q3: What is unsupervised learning? List some examples of unsupervised learning.

# Unsupervised learning is a type of machine learning where the algorithm is trained on an unlabeled dataset.
# Unlike supervised learning, there are no provided output labels during the training phase.
# The goal of unsupervised learning is to identify patterns, structures, or relationships 
# within the data without any specific guidance or predefined categories.
# Examples of unsupervised learning: Clustering, Anomaly Detection, Dimensionality Reduction, 
# Market Basket Analysis, Document Clustering, Image Compression, Social Network Analysis.
```

```python
# Q6: What is train, test and validation split? Explain the importance of each term.

# In machine learning, the process of training a model involves using a dataset to teach the model
# to make predictions or learn patterns from the data. However, it is essential to evaluate how well 
# the model performs on new, unseen data to avoid overfitting (memorizing the training data) 
# or underfitting (failing to capture patterns). To achieve this, the dataset is typically split 
# into three subsets: the training set, the validation set, and the test set.

# Training Set : The training set is the portion of the dataset used to train the machine learning model.
# It contains labeled data, where both the input features and their corresponding output labels 
# are provided to the model. During training, the model learns to map the input data to the correct 
# output labels by adjusting its internal parameters. The goal is to minimize the prediction error
# on the training set and generalize well to unseen data.

# Validation Set : The validation set is used to fine-tune the hyperparameters of the model and 
# evaluate its performance during the training process. Hyperparameters are settings that are not 
# learned by the model but are set before training, such as the learning rate, number of hidden layers,
# or the number of neurons in each layer. By using the validation set, we can experiment with 
# different hyperparameter values and choose the ones that result in the best performance on the validation set.

# The validation set helps to detect overfitting. As the model trains on the training set, it may start 
# to perform very well on the training data but fail to generalize to new data (i.e., the validation set).
# By evaluating the model on the validation set, we can identify whether the model is overfitting 
# or if we need to adjust the hyperparameters to achieve better generalization.

# Test Set : The test set is a separate portion of the dataset that is not used during training 
# or hyperparameter tuning. It serves as an unbiased evaluation set to assess the model's performance 
# after the training and fine-tuning phases are complete. The test set provides an estimation of 
# how well the model will perform in the real world on new, unseen data.

# Importance of each split : Training Set Importance: The training set is crucial as it is used to 
# teach the model and adjust its parameters to learn from the data. The model's ability to generalize 
# and make accurate predictions on unseen data depends on how well it learns from the training set.

# Validation Set Importance: The validation set is essential for hyperparameter tuning and model selection.
# It helps us choose the best hyperparameters that yield good generalization to unseen data and avoid overfitting.

# Test Set Importance: The test set provides an unbiased assessment of the model's performance on new,
# unseen data. It serves as a final evaluation to determine how well the model is expected to perform
# in real-world scenarios.
```

```python
# Q7: How can unsupervised learning be used in anomaly detection?

# Unsupervised learning is commonly used in anomaly detection because it can identify patterns 
# or structures in data without the need for labeled examples of anomalies. Anomalies, 
# also known as outliers, are data points that deviate significantly from the normal behavior 
# or the majority of the data. Anomaly detection with unsupervised learning involves the following approaches:

# Clustering-Based Anomaly Detection : In this approach, unsupervised learning algorithms are 
# used to group similar data points together into clusters. Data points that do not belong to 
# any well-defined cluster or are isolated from other clusters are considered anomalies. 
# One common algorithm used for clustering-based anomaly detection is DBSCAN 
#(Density-Based Spatial Clustering of Applications with Noise). DBSCAN identifies dense regions of 
# data points and marks isolated or low-density points as anomalies.

# Autoencoders : Autoencoders are a type of neural network used for dimensionality reduction.
# They consist of an encoder and a decoder, and the network is trained to reconstruct its input
# data from a lower-dimensional representation called the latent space. During training, 
# the autoencoder learns to capture the regular patterns in the data. When new data is passed 
# through the trained autoencoder, data points that cannot be accurately reconstructed or 
# have a high reconstruction error are considered anomalies.

# One-Class SVM (Support Vector Machine) : One-Class SVM is a variation of the traditional 
# SVM algorithm designed for anomaly detection. It learns the boundary that encloses the majority of 
# the data points, assuming that the majority of the data belongs to the normal class. 
# Data points that fall outside this boundary are considered anomalies.

# Isolation Forest : Isolation Forest is an ensemble-based method that constructs random decision 
# trees to isolate anomalies efficiently. It works by randomly selecting features and partitioning
# the data into subsets. The process is repeated recursively until each data point is isolated 
# or grouped with other anomalies.

# Density Estimation : In this approach, unsupervised learning algorithms estimate the probability 
# density function of the data. Anomalies are identified as data points with low probability density, 
# indicating that they are far from the dense regions where most of the data resides.
```

```python
# Q8: List down some commonly used supervised learning algorithms and unsupervised learning algorithms.

# Supervised Learning Algorithms : 
# Linear Regression: A simple algorithm used for regression tasks to 
# model the relationship between input features and continuous output variables.
# Logistic Regression: Used for binary classification tasks, logistic regression predicts 
# the probability of an instance belonging to a particular class.
# Decision Trees: Decision trees are tree-like structures used for both classification and 
# regression tasks. They split data based on features to make decisions.
# Random Forest: An ensemble method that combines multiple decision trees to improve accuracy 
# and reduce overfitting.
# Support Vector Machines (SVM): SVM is used for both classification and regression tasks.
# It finds the optimal hyperplane that separates classes in the feature space.
# K-Nearest Neighbors (KNN): A simple algorithm that classifies a data point based on 
# the majority class of its k-nearest neighbors.
# Naive Bayes: A probabilistic algorithm based on Bayes' theorem used for classification tasks,
# particularly in natural language processing.
# Neural Networks: Deep learning models composed of layers of artificial neurons, 
# used for a wide range of tasks, including image and speech recognition, language translation, etc.

# Unsupervised Learning Algorithms:
# K-Means Clustering: An algorithm that partitions data into k clusters based on similarity, 
# where each cluster is represented by its centroid.
# Hierarchical Clustering: A method that creates a hierarchical representation of clusters, 
# forming a tree-like structure (dendrogram).
# DBSCAN (Density-Based Spatial Clustering of Applications with Noise): A density-based clustering algorithm 
# that identifies clusters based on areas of high data density.
# PCA (Principal Component Analysis): A dimensionality reduction technique used to transform data 
# into a lower-dimensional space while preserving the most important information.
# Autoencoders: Neural network architectures used for unsupervised representation learning and feature extraction.
# Isolation Forest: An ensemble-based algorithm that efficiently identifies anomalies by isolating 
# them using random decision trees.
# Gaussian Mixture Models (GMM): A probabilistic model that assumes data points are generated 
# from a mixture of several Gaussian distributions.
# Self-Organizing Maps (SOM): An artificial neural network that learns to map high-dimensional
# data onto a lower-dimensional grid, useful for visualizing complex data.

```

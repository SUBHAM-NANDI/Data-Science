```python
# Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.

# Probability Mass Function (PMF) : The Probability Mass Function (PMF) is used for discrete random variables. 
# It gives the probability of a discrete random variable taking on a specific value. 
# The PMF provides a mapping from each possible value of the random variable to its associated probability.
# For a discrete random variable X, the PMF is denoted by P(X = x), where x is a specific value in the range of X. 
# The PMF satisfies two properties:
# P(X = x) ≥ 0 for all x.
# The sum of the probabilities over all possible values of X is equal to 1.
# Example of PMF:
# Consider rolling a fair six-sided die. The random variable X represents the outcome of the roll. 
# Since there are only six possible outcomes (1, 2, 3, 4, 5, or 6), X is a discrete random variable. The PMF for X is as follows:
# P(X = 1) = 1/6
# P(X = 2) = 1/6
# P(X = 3) = 1/6
# P(X = 4) = 1/6
# P(X = 5) = 1/6
# P(X = 6) = 1/6

# Probability Density Function (PDF):
# The Probability Density Function (PDF) is used for continuous random variables. 
# Unlike the PMF, which deals with individual probabilities of discrete outcomes, 
# the PDF gives the relative likelihood of the random variable falling within a certain range of values. 
# The area under the PDF curve over a range represents the probability of the random variable falling within that range.
# For a continuous random variable X, the PDF is denoted by f(x). The PDF satisfies two properties:
# f(x) ≥ 0 for all x.
# The area under the PDF curve over the entire range is equal to 1.
# Example of PDF:
# Consider the standard normal distribution, which has a bell-shaped curve with a mean of 0 and a standard deviation of 1. 
# The PDF for this distribution is given by:
# f(x) = (1 / √(2π)) * e^(-(x^2) / 2)
# In this example, the PDF f(x) represents the probability density at each point 'x' along the curve. 
# The probability of the random variable X falling within a specific range is calculated by integrating the PDF over that range. 
# For example, to find the probability of X being between -1 and 1, you would calculate the area under the curve of the PDF between -1 and 1.
```

```python
# Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?

# The Cumulative Density Function (CDF) is a function used in probability theory and statistics to describe 
# the cumulative probability of a random variable being less than or equal to a specific value. 
# It provides a way to determine the probability of a random variable falling within a certain range.
# Definition of CDF : For a random variable X, the CDF is denoted by F(x) and is defined as follows:
# F(x) = P(X ≤ x)
# In words, the CDF at a given point 'x' represents the probability that the random variable X takes on a value less than or equal to 'x'.

# Example of CDF : Let's consider the rolling of a fair six-sided die. The random variable X represents the outcome of the roll. 
# Since there are only six possible outcomes (1, 2, 3, 4, 5, or 6), X is a discrete random variable.
# The CDF for X in this case is as follows:
# F(x) = P(X ≤ x)
# For x = 1: F(1) = P(X ≤ 1) = 1/6 (since there is one outcome less than or equal to 1, which is 1 itself).
# For x = 2: F(2) = P(X ≤ 2) = 2/6 = 1/3 (two outcomes, 1 and 2, are less than or equal to 2).
# For x = 3: F(3) = P(X ≤ 3) = 3/6 = 1/2 (three outcomes, 1, 2, and 3, are less than or equal to 3).
# For x = 4: F(4) = P(X ≤ 4) = 4/6 = 2/3 (four outcomes, 1, 2, 3, and 4, are less than or equal to 4).
# For x = 5: F(5) = P(X ≤ 5) = 5/6 (five outcomes, 1, 2, 3, 4, and 5, are less than or equal to 5).
# For x = 6: F(6) = P(X ≤ 6) = 1 (all six outcomes are less than or equal to 6).

# The CDF is used for several reasons:
# Calculating Probabilities: The CDF allows us to find the probability that a random variable falls within a certain range. 
# For example, the probability of rolling a value less than or equal to 3 on a fair six-sided die can be found using the CDF.
# Understanding Distribution Characteristics: By examining the CDF, we can understand the characteristics of the distribution, 
# such as its shape, symmetry, and range of values.
# Comparing Distributions: The CDF allows us to compare different distributions and analyze their similarities and differences.
# Simulation and Modeling: The CDF is used in simulations and statistical modeling to generate random samples from a given distribution.
```

```python
# Q3: What are some examples of situations where the normal distribution might be used as a model?
# Explain how the parameters of the normal distribution relate to the shape of the distribution.

# The normal distribution, also known as the Gaussian distribution, is widely used as a model in various real-world situations 
# due to its versatility and prevalence in nature. Some examples of situations where the normal distribution might be used as a model include:
# height of individuals,measurement errors, exam score, financial markets, IQ scores, errors in manufacturing, etc.

# Parameters of the Normal Distribution : The normal distribution is fully characterized by two parameters: the mean (μ) and the standard deviation (σ).
# Mean (μ): The mean represents the central tendency or the average of the distribution. 
# It is the point around which the data cluster, and it determines the position of the peak (mode) of the bell-shaped curve.
# Standard Deviation (σ): The standard deviation measures the spread or dispersion of the data around the mean. 
# A larger standard deviation results in a broader and flatter curve, indicating higher variability in the data. 
# Conversely, a smaller standard deviation leads to a narrower and taller curve, indicating less variability.
# The combination of the mean and standard deviation determines the shape of the normal distribution. 
# A larger standard deviation leads to a flatter distribution, spreading the data over a wider range of values. 
# A smaller standard deviation results in a more peaked distribution, with the data concentrated closely around the mean.
```

```python
# Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli Distribution and Binomial Distribution?

# Bernoulli Distribution: The Bernoulli distribution is a discrete probability distribution that models a random experiment 
# with only two possible outcomes, usually labeled as success (1) and failure (0). 
# It is named after the Swiss mathematician Jacob Bernoulli. The distribution is characterized by a single parameter 'p', 
# which represents the probability of success in a single trial.

# Example of Bernoulli Distribution:
# A classic example of the Bernoulli distribution is modeling the outcome of flipping a fair coin. 
# In this case, the random variable 'X' can take two values: 1 (success: heads) 
# with a probability of 0.5, and 0 (failure: tails) with a probability of 0.5. If the coin is biased, 
# the probability of success (heads) would be different from 0.5, making it a non-fair Bernoulli distribution.

# Difference between Bernoulli Distribution and Binomial Distribution:
# 1.Number of Trials:
# - Bernoulli Distribution: It is used for a single trial or a sequence of independent, identical trials where each trial has two possible outcomes.
# - Binomial Distribution: It is used for a fixed number of independent Bernoulli trials, where each trial can have two outcomes, 
# and we are interested in the number of successes in a given number of trials.
# 2.Parameters:
# - Bernoulli Distribution: It has a single parameter 'p', which is the probability of success in a single trial.
# - Binomial Distribution: It has two parameters: 'n', the number of trials, and 'p', the probability of success in a single trial.
# 3.Number of Possible Outcomes:*
# - Bernoulli Distribution: It has two possible outcomes: success (1) and failure (0).
# - Binomial Distribution: It deals with discrete random variables representing the number of successes (0 to 'n') in 'n' trials.
# 4.Probability Mass Function (PMF):
# - Bernoulli Distribution: The PMF of the Bernoulli distribution describes the probabilities of the two possible outcomes (success and failure).
# - Binomial Distribution: The PMF of the Binomial distribution describes the probabilities of getting exactly 'k' successes in 'n' trials.
```

```python
# Q8: What is the z score? State the importance of the z score.

# The z-score, also known as the standard score, is a statistical measure that quantifies 
# the number of standard deviations a data point is from the mean of a dataset. It is used to standardize data and allows
# for comparisons between different datasets that may have different scales and units of measurement.

# Importance of the z-score:
# The z-score is essential for several reasons:
# Standardization: It standardizes data by transforming it into a common scale with a mean of 0 and a standard deviation of 1. 
# This makes it easier to compare and analyze data from different distributions.
# Outlier Detection: The z-score helps identify outliers in a dataset. Data points with z-scores significantly 
# greater than or less than zero are considered outliers, as they lie far from the mean of the distribution.
# Probability Calculation: The z-score is used to calculate probabilities in the standard normal distribution (z-distribution) table.
# The z-distribution is a special case of the normal distribution with a mean of 0 and a standard deviation of 1.
# Data Normality Testing: The z-score is used in normality tests to assess whether a dataset follows a normal distribution.
# If the z-scores of data points deviate significantly from zero, it may suggest that the data is not normally distributed.
# Statistical Analysis: Z-scores are widely used in statistical analysis, hypothesis testing, and confidence interval estimation.
# Data Transformation: Z-scores are used in various data transformation techniques to achieve certain 
# statistical properties or assumptions in regression and other analyses.
```

```python
# Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

# The Central Limit Theorem (CLT) is a fundamental concept in statistics that states that when independent random variables are added,
# their sum tends to follow a normal distribution, regardless of the distribution of the individual random variables. 
# This theorem is one of the most important and widely used principles in statistics, 
# as it allows us to make inferences about population parameters based on sample statistics.

# Significance of the Central Limit Theorem:
# The Central Limit Theorem has significant implications in statistics and data analysis:
# Sampling Distributions: The CLT justifies the use of normal distribution-based sampling distributions for various sample statistics,
# such as the sample mean and sample proportion. This allows us to make probabilistic inferences about population parameters based on sample data.
# Hypothesis Testing and Confidence Intervals: The CLT is the foundation for many hypothesis tests and confidence intervals. 
# It allows us to construct confidence intervals and perform hypothesis tests on population parameters, 
# such as the mean and proportion, using the standard normal distribution.
# Approximation of Distributions: For large sample sizes, the CLT allows us to approximate the distribution
# of sample statistics (e.g., sample mean) as normal, even if the underlying population distribution is not normal. 
# This approximation is extremely useful in practical applications.
# Statistical Inference: The CLT is at the core of statistical inference, enabling researchers to draw conclusions 
# about populations based on information from representative samples.
# Generalization to Real-World Data: The theorem is widely applicable in real-world scenarios, 
# as many real-world data tend to exhibit a normal distribution due to the combined effects of numerous independent factors.
```

```python
# Q10: State the assumptions of the Central Limit Theorem.

# Independence: The random variables in the sample must be independent of each other. 
# This means that the outcome of one random variable does not affect the outcome of another. 
# Independence is crucial for the CLT because it ensures that the combined effect of multiple random variables leads to a normal distribution.

# Identically Distributed: The random variables in the sample should be identically distributed. 
# This means that each random variable in the sample comes from the same population and follows the same probability distribution. 
# The assumption of identical distribution is essential to ensure that the combined sample is representative of the population.

# Sample Size: The sample size should be sufficiently large. While there is no strict rule on the minimum sample size, 
# as the sample size increases, the sample mean tends to follow a normal distribution even if the individual population distribution is not normal. 
# As a general guideline, a sample size of 30 or larger is often considered sufficient for the CLT to hold.

# Finite Variance: The individual random variables should have finite variance. If the variance is infinite, the CLT may not apply.

# Population Distribution: The CLT does not require the population distribution to be normal. 
# The theorem works even if the underlying population follows a different distribution. 
# However, the more the population distribution deviates from normality, the larger the sample size needed to achieve a normal sampling distribution.

# No Heavy Tails: The CLT is most reliable when dealing with distributions without heavy tails or extreme outliers. 
# In cases where the population distribution has heavy tails, the CLT may still apply, but larger sample sizes may be required for reliable approximations.
```
